# Wide Audience Plots

A reprocessing some of the first figures from Hafen+2018 into a form that's more useful for a broad audience.

## The CGM is Massive

```python
import verdict
```

```python
data, attrs = verdict.Dict.from_hdf5( '../data/halo_mass_content_FIRE2_snum600.hdf5' )
```

```python
data['M']['stars']['m_gal']
```

### Baryon Mass Fraction, Black Background

```python
plt.style.use('dark_background')

fig = plt.figure( figsize=(14,10), )
ax = plt.gca()

ax.scatter(
    data['M']['stars']['m_gal'],
    ( data['M']['gas']['m_cgm'] / data['M']['total']['m_halo'] ),
    s = 200,
    color = 'w',
)

# Tweak Axes
ax.set_ylim( 0, 1 )
ax.set_xscale( 'log' )

plt.tick_params( axis='both', right=True, labelright=True,  labelsize=32, size=20 )
plt.tick_params( axis='both', which='minor', size=14, width=2 )

# Labels
ax.set_xlabel( 'Galaxy Stellar Mass ($M_\odot$)', fontsize=28, fontname='Helvetica', )
ax.set_ylabel(
    'Fraction of\nBaryonic Mass\nin CGM',
    fontsize=28, 
    fontname='Helvetica',  
    rotation='horizontal',
    verticalalignment='center',
    labelpad=110,
)
ax.set_title(
    "The CGM contains much of the\nregular matter in a galaxy's halo",
    fontsize=38,
    y = 1.05,
    fontname='Helvetica',
)
ax.annotate(
    xy=(0.01,0.01),
    s='z = {:.3g}'.format( attrs['redshift'] ),
    xycoords='axes fraction',
    fontsize=36,
    ha = 'left',
    va = 'bottom',
)
```

### Gas Mass Fraction, Black Background

```python
plt.style.use('dark_background')

fig = plt.figure( figsize=(14,10), )
ax = plt.gca()

ax.scatter(
    data['M']['stars']['m_gal'],
    ( data['M']['gas']['m_cgm'] / data['M']['gas']['m_halo'] ),
    s = 200,
    color = 'w',
)

# Tweak Axes
ax.set_ylim( 0, 1 )
ax.set_xscale( 'log' )

plt.tick_params( axis='both', right=True, labelright=True,  labelsize=32, size=20 )
plt.tick_params( axis='both', which='minor', size=14, width=2 )

# Labels
ax.set_xlabel( 'Galaxy Stellar Mass ($M_\odot$)', fontsize=28, fontname='Helvetica', )
ax.set_ylabel(
    'Fraction of\nGas Mass\nin CGM',
    fontsize=28, 
    fontname='Helvetica',  
    rotation='horizontal',
    verticalalignment='center',
    labelpad=100,
)
ax.set_title(
    "The CGM contains the majority of \nthe gas in a galaxy's halo",
    fontsize=38,
    y = 1.05,
    fontname='Helvetica',
)
ax.annotate(
    xy=(0.01,0.01),
    s='z = {:.3g}'.format( attrs['redshift'] ),
    xycoords='axes fraction',
    fontsize=36,
    ha = 'left',
    va = 'bottom',
)
```
